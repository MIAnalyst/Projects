| Type of Kurtosis      | Meaning                                 |
| --------------------- | --------------------------------------- |
| **3** (Normal)        | Normal bell curve (moderate tails)      |
| **> 3** (Leptokurtic) | Sharp peak, heavy tails (more outliers) |
| **< 3** (Platykurtic) | Flat peak, light tails (fewer outliers) |

Tukey's Rule (1.5 * IQR Rule) is a common statistical method for outlier detection, used widely in data analysis by both academics and Big Tech companies.
- Determine Outlier Thresholds:
    - Lower Bound: Q1 - 1.5 * IQR
    - Upper Bound: Q3 + 1.5 * IQR
- Outliers:
    - Any data point below the lower bound or above the upper bound is considered an outlier.

```python
import pandas as pd
import numpy as np
from scipy.stats import entropy
from phik import phik_matrix

def datType(df,c):
    if df[c].nunique() > 20 and df[c].dtype.kind in "iufc":
        return "Continuous"
    elif df[c].nunique() < 20 and df[c].dtype.kind in "iufc":
        return "Categorical"
    elif df[c].dtype==object:
        return "Text"
    elif pd.api.types.is_datetime64_any_dtype(df[c]):
        listt.append("Date")
    else:
        return "Other"

def DatasetMetadata(df, toPrint=True):
    summary_data = []
    #Since the memory usage of columns doesn't change during the loop, there's no need to recalculate it inside the loop.
    memUsage = list(df.memory_usage(deep=True) / (1024 * 1024))[1:]

    for idx, c in enumerate(df.columns):
        col = df[c]
        dtype = col.dtype
        mem = memUsage[idx]
        uniq = col.nunique()
        nulls = col.isnull().sum()
        empty = col.astype(str).str.strip().eq("").sum()
        mode_val = col.mode()[0] if not col.mode().empty else "N/A"
        ent = round(entropy(col.value_counts(normalize=True), base=2), 2)
        maxent = round(np.log2(uniq), 2) if uniq > 0 else 0

        # Defaults for all columns
        mean = std = min_val = max_val = skew = kurt = q1 = q2 = q3 = outliers = "N/A"

        if dtype.kind in "iufc":
            mean = round(col.mean(), 2)
            std = round(col.std(), 2)
            min_val = col.min()
            max_val = col.max()
            q1 = col.quantile(0.25)
            q2 = col.quantile(0.5)
            q3 = col.quantile(0.75)
            iqr = q3 - q1
            lower = q1 - 1.5 * iqr
            upper = q3 + 1.5 * iqr
            outliers = ((col < lower) | (col > upper)).sum()
            if uniq > 20:
                skew = round(col.skew(), 2)
                kurt = round(col.kurtosis(), 2)

        vartype = datType(df,c)

        summary_data.append([
            c, str(dtype), mem, vartype, uniq, nulls, empty, mode_val,
            mean, std, min_val, q2, max_val, skew, kurt, q1, q2, q3, outliers, ent, maxent
        ])

    SummaryDF = pd.DataFrame(summary_data, columns=[
        "CName", "DType", "UsageMB", "VarType", "Cardinality", "Null*", "Empty*", "Mode",
        "Mean", "Std", "Min", "Median", "Max", "Skewness", "Kurtosis", "Q1", "Q2/Median", "Q3",
        "Outliers", "Entropy", "Max Entropy"
    ])

    if toPrint:
        print(f"The dataset includes {df.shape[1]} columns and {df.shape[0]} rows")
        print(f"Dataset has {df.duplicated().sum()} duplicate rows")
        print(f"Memory Usage of the DataFrame: {df.memory_usage(deep=True).sum() / (1024 * 1024):.2f} MB")
        print("--------------------------")
        print("Phik Correlation Matrix")
        numeric = [c for c in df.columns if df[c].dtype.kind in "iufc"]
        corr_matrix = df.phik_matrix(interval_cols=numeric)
        print(corr_matrix)
        print("--------------------------")
        print("First rows")
        print(df.head())
        print("--------------------------")

    return SummaryDF
```

```python
def datType(c):
    if c.nunique() > 20 and c.dtype.kind in "iufc":
        return "Continuous"
    elif c.nunique() < 20 and c.dtype.kind in "iufc":
        return "Categorical"
    elif c.dtype==object:
        return "Text"
    elif pd.api.types.is_datetime64_any_dtype(c):
        return "Date"
    else:
        return "Other"

def meta(c):
    dtype = c.dtype
    mem = round(c.memory_usage(deep=True) / (1024 * 1024),2)
    uniq = c.nunique()
    nulls = c.isnull().sum()
    empty = c.astype(str).str.strip().eq("").sum()
    return dtype,mem,uniq,nulls,empty

def stats(c):
    mode_val = c.mode()[0] if not c.mode().empty else "N/A"
    ent = round(entropy(c.value_counts(normalize=True), base=2), 2)
    maxent = round(np.log2(c.nunique()), 2) if c.nunique() > 0 else 0
    # Defaults for all columns
    mean = std = min_val = max_val = skew = kurt = q1 = q2 = q3 = outliers = "N/A"
    if c.dtype.kind in "iufc":
            mean = round(c.mean(), 2)
            std = round(c.std(), 2)
            min_val = c.min()
            max_val = c.max()
            q1 = c.quantile(0.25)
            q2 = c.quantile(0.5)
            q3 = c.quantile(0.75)
            iqr = q3 - q1
            lower = q1 - 1.5 * iqr
            upper = q3 + 1.5 * iqr
            outliers = ((c < lower) | (c > upper)).sum()
            if c.nunique() > 20:
                skew = round(c.skew(), 2)
                kurt = round(c.kurtosis(), 2)
    return mode_val,ent,maxent,mean ,std ,min_val , max_val ,skew , kurt , q1 , q2 , q3 , outliers

def DatasetMetadata(df, toPrint=True):
    summary_data = []
    #Since the memory usage of columns doesn't change during the loop, there's no need to recalculate it inside the loop.
    memUsage = list(df.memory_usage(deep=True) / (1024 * 1024))[1:]

    for  c in df.columns:
        col = df[c]
        dtype,mem,uniq,nulls,empty=meta(col)

        mode_val,ent,maxent,mean ,std ,min_val , max_val ,skew , kurt , q1 , q2 , q3 , outliers=stats(col)
        vartype = datType(col)

        summary_data.append([
            c, str(dtype), mem, vartype, uniq, nulls, empty, mode_val,
            mean, std, min_val, q2, max_val, skew, kurt, q1, q2, q3, outliers, ent, maxent
        ])

    SummaryDF = pd.DataFrame(summary_data, columns=[
        "CName", "DType", "UsageMB", "VarType", "Cardinality", "Null*", "Empty*", "Mode",
        "Mean", "Std", "Min", "Median", "Max", "Skewness", "Kurtosis", "Q1", "Q2/Median", "Q3",
        "Outliers", "Entropy", "Max Entropy"
    ])

    if toPrint:
        print(f"The dataset includes {df.shape[1]} columns and {df.shape[0]} rows")
        print(f"Dataset has {df.duplicated().sum()} duplicate rows")
        print(f"Memory Usage of the DataFrame: {df.memory_usage(deep=True).sum() / (1024 * 1024):.2f} MB")
        print("--------------------------")
        print("Phik Correlation Matrix")
        numeric = [c for c in df.columns if df[c].dtype.kind in "iufc"]
        corr_matrix = df.phik_matrix(interval_cols=numeric)
        print(corr_matrix)
        print("--------------------------")
        print("First rows")
        print(df.head())
        print("--------------------------")

    return SummaryDF
```
